## 4. Time Range Analysis

This is a temporal dataset - we need to understand what years are available.

**Critical Question:** Do we have 10 complete years of data?

```python
# ============================================
# IDENTIFY THE YEAR COLUMN
# ============================================
# The year column is typically named 'Survey Year' in EADA data
# Let's find it and analyze the time range

# Check for common year column names
year_candidates = ['Survey Year', 'Year', 'YEAR', 'year']
year_col = None

for candidate in year_candidates:
    if candidate in df.columns:
        year_col = candidate
        break

if year_col:
    print(f"âœ… Found year column: '{year_col}'")
    print(f"\nðŸ“… YEAR RANGE:")
    print(f"   Minimum Year: {df[year_col].min()}")
    print(f"   Maximum Year: {df[year_col].max()}")
    print(f"   Total Years: {df[year_col].nunique()}")
    print(f"\nðŸ“Š ROWS PER YEAR:")
    print(df[year_col].value_counts().sort_index())
else:
    print("âš ï¸ Could not find a year column")
    print("Available columns:", list(df.columns))
```

---
## 5. Institution Analysis

Each row represents one institution in one year. Let's understand:
- How many unique institutions?
- Do all institutions have data for all years?

```python
# ============================================
# INSTITUTION COUNT
# ============================================
# UNITID is the unique identifier for each institution
# It's a federal ID assigned to each college/university

if 'UNITID' in df.columns:
    n_institutions = df['UNITID'].nunique()
    print(f"ðŸ« UNIQUE INSTITUTIONS: {n_institutions:,}")
    
    # Check if we have balanced data (same institutions each year)
    expected_rows = n_institutions * df[year_col].nunique()
    actual_rows = len(df)
    
    print(f"\nðŸ“Š DATA BALANCE CHECK:")
    print(f"   Expected rows (if all institutions Ã— all years): {expected_rows:,}")
    print(f"   Actual rows: {actual_rows:,}")
    
    if actual_rows == expected_rows:
        print(f"   âœ… Data is balanced - all institutions have all years")
    else:
        print(f"   âš ï¸ Some institutions may be missing some years")
        print(f"   Missing: {expected_rows - actual_rows:,} rows")
```

```python
# ============================================
# SAMPLE INSTITUTIONS
# ============================================
# Let's look at a few institution names to understand the scope

if 'Institution Name' in df.columns:
    print("ðŸ« SAMPLE INSTITUTIONS (first 20):")
    print("=" * 60)
    sample_institutions = df['Institution Name'].drop_duplicates().head(20)
    for i, name in enumerate(sample_institutions, 1):
        print(f"{i:2}. {name}")
```

## 6. Key Financial Columns

For our analysis, the most important columns are:
- **Grand Total Revenue**: Total athletic revenue
- **Grand Total Expenses**: Total athletic expenses

These will be the foundation of our financial trajectory prediction.

```python
# ============================================
# KEY FINANCIAL COLUMNS - BASIC STATISTICS
# ============================================
# These are the core financial metrics we'll use for prediction

financial_cols = ['Grand Total Revenue', 'Grand Total Expenses']

print("ðŸ’° KEY FINANCIAL METRICS - SUMMARY STATISTICS:")
print("=" * 60)

for col in financial_cols:
    if col in df.columns:
        print(f"\nðŸ“Š {col}:")
        print(f"   Count:   {df[col].count():,}")
        print(f"   Missing: {df[col].isna().sum():,}")
        print(f"   Min:     ${df[col].min():,.0f}")
        print(f"   Max:     ${df[col].max():,.0f}")
        print(f"   Mean:    ${df[col].mean():,.0f}")
        print(f"   Median:  ${df[col].median():,.0f}")
```

---
## 7. Missing Values Analysis

Before proceeding, we need to understand data quality:
- Which columns have missing values?
- How severe is the missing data problem?

```python
# ============================================
# MISSING VALUES SUMMARY
# ============================================
# Calculate missing values for each column
# This helps us understand data quality before analysis

missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)

# Create a summary DataFrame
missing_df = pd.DataFrame({
    'Column': missing.index,
    'Missing Count': missing.values,
    'Missing %': missing_pct.values
})

# Show only columns with missing values
missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values('Missing Count', ascending=False)

print("âš ï¸ COLUMNS WITH MISSING VALUES:")
print("=" * 60)

if len(missing_df) > 0:
    for _, row in missing_df.iterrows():
        print(f"{row['Column']:<45} {row['Missing Count']:>6,} ({row['Missing %']:>5}%)")
else:
    print("âœ… No missing values found!")
```

---
## 8. Classification/Division Analysis

NCAA institutions are classified into divisions (D1, D2, D3). This is important because:
- D1 schools typically have much larger athletic budgets
- Financial patterns may differ by division

```python
# ============================================
# DIVISION/CLASSIFICATION ANALYSIS
# ============================================
# The classification column tells us what division each school belongs to

if 'Classification Name' in df.columns:
    print("ðŸ† INSTITUTIONS BY CLASSIFICATION:")
    print("=" * 60)
    
    # Count unique institutions by classification (not rows)
    class_counts = df.groupby('Classification Name')['UNITID'].nunique().sort_values(ascending=False)
    
    for classification, count in class_counts.items():
        print(f"{classification:<50} {count:>5} institutions")
```

---
## 9. State Distribution

Let's see how institutions are distributed across states.

```python
# ============================================
# STATE DISTRIBUTION
# ============================================
# Count unique institutions by state

if 'State CD' in df.columns:
    print("ðŸ“ TOP 15 STATES BY NUMBER OF INSTITUTIONS:")
    print("=" * 40)
    
    state_counts = df.groupby('State CD')['UNITID'].nunique().sort_values(ascending=False).head(15)
    
    for state, count in state_counts.items():
        print(f"{state:<10} {count:>5} institutions")
```
